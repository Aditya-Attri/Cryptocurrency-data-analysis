```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
from datetime import datetime, timedelta
import matplotlib.dates as mdates
```

```python
data = pd.read_csv("crypto-markets.csv")
data.head()
```

```python
data["date"] = pd.to_datetime(data["date"], format="%Y-%m-%d")
print("Latest crypto data")
latest_data = data[data["date"] == max(data["date"])]
latest_data.head()
```

```python
x = 10
start_data = pd.DataFrame({'start_date' : data.groupby(["name", "ranknow"])['date'].min()}).reset_index()
start_data.sort_values(['start_date']).head(x)
```

```python
start_data.sort_values(['start_date']).tail(x)
```

```python
latest_data[latest_data['ranknow'] <= x].groupby('ranknow').name.unique()
```

```python
latest_data.sort_values(by=['market'], ascending=[False])
latest_data[['name', 'market']].head(x)
```

```python
#MARKET_VALUES
name = latest_data['name'].unique()
currency = []
market_value = []
x_currency = name[:x]
for i, cn in enumerate(x_currency):
    filtered = latest_data[(latest_data['name'] == str(cn))]
    currency.append((cn))
    market_value.append(filtered['market'].values[0])

f, ax = plt.subplots(figsize = (20, 8))
g = sns.barplot(y=currency, x = market_value, palette=sns.cubehelix_palette(x, reverse=True))
plt.title("Top X Cryptocurrencies market value")
ax.set_xticklabels(ax.get_xticks())
fig = plt.gcf()
plt.show()
```

```python
currency = []
volume = []
x_currencies = name[:x]
for i, cn in enumerate(x_currencies):
    filtered = latest_data[(latest_data['name']==str(cn))]
    currency.append((cn))
    volume.append(filtered['volume'].values[0])

f, ax = plt.subplots(figsize=(20, 8))
g = sns.barplot(y=currency, x=volume, palette=sns.cubehelix_palette(x, reverse=True))
plt.title("Top X Cryptocurrencies volume")
ax.set_xticklabels(ax.get_xticks())
fig = plt.gcf()
plt.show()
```

```python
from mplfinance.original_flavor import candlestick_ohlc
```

```python
#Candlestick Charts 
rank = 1
months = 6

name = data[data.ranknow == rank].iloc[-1]['name']
filtered_data = data[(data['ranknow'] == rank) & (data['date'] > (max(data['date']) - timedelta(days=30*months)))]
OHLCfiltered_data = filtered_data[['date','open','high','low','close']]
OHLCfiltered_data['date'] = mdates.date2num(OHLCfiltered_data['date'].dt.date)

f,ax=plt.subplots(figsize=(15,11))
ax.xaxis_date()
candlestick_ohlc(ax, OHLCfiltered_data.values, width=0.5, colorup='g', colordown='r',alpha=0.75)

plt.xlabel("Date")
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()
plt.title(name + " price")
plt.ylabel("Price")
plt.show()
```
